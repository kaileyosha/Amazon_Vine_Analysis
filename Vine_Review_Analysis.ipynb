```python
#import dependencies
import os 
import pandas as pd 

os.getcwd()

```

```python
#read in vine review 
df = pd.read_csv("vine_table.csv")

#check
df.head()

df.info()
```

```python
#more than 20 votes 

new_df = df.loc[(df["total_votes"] >= 20)]

#check
#original error: forgot brackets around (df[]) so it came back as unhashable 
new_df.head()
```

```python
#filter based on helpful/total being greater than 50%

#first making a new variable 

new_df["percent"] = (new_df["helpful_votes"]/new_df["total_votes"]) * 100

#next make data frame where that is greater than 50 

third_df = new_df.loc[(new_df["percent"] > 50)]

#check

third_df.head() 
```

```python
#vine program table 

vine_df = third_df.loc[(third_df["vine"] == "Y")]

#check
vine_df.head()
```

```python
#not vine table 

not_vine_df = third_df.loc[(third_df["vine"] == "N")]

not_vine_df.head()
```
